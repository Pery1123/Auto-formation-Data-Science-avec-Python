## Partie 1:
### Créer une fonction initialisation() qui prend en parmètre m (nombre de lignes) et n (nombre de colonnes), cette fonction doit ensuite retourner une matrice aléatoire  (m, n+1) avec une colonne biais (remplie de "1") tout à droite

```python
# Premiere Version de la fonction initialisation()

def initialisation(m, n):
    # Création de la matrice aléatoire
    matrice = np.random.randn(m, n)
    
    # Une colonne de biais
    biais = np.ones((n, 1))
    
    # Rédimensionner la matrice aléatoire 
    matrice = matrice.reshape((matrice.shape[1], m))
    
    # Retourner la matrice qui contient la concatenation de la matrice normale et la colonne des biais
    return np.concatenate((matrice, biais), axis=1)
    
```

```python
print(initialisation(3, 4))
```

## Correction

```python
# Version améliorée de la fonction initialization()

def initialization(m, n):
    
    # Création de la mtrice aléatoire
    X = np.random.randn(m, n)
    # Assembler cette matrice avec une colonne remplie de 1 tout à droite
    X = np.concatenate((X, np.ones((X.shape[0], 1))), axis=1)
    
    return X
```

```python
initialization(3, 4)
```

### LE SLICILNG AVEC PAS : A[début:fin:pas, début:fin:pas]

```python
# Création d'une matrice de dimension 5 remplie de 0
C = np.zeros((5, 5))
C
```

```python
# Remplacer tous les éléments du tableau par 1 en sautant un pas de 2, 
# C'est-à-dire ne pas toucher la 2ème et 4ème ligne et la 2ème et 4ème colonne
C[::2, ::2] = 1
C
```

### NB : En Machine Learning et Data Science, on ne travaille pas avec un si grand pas, car cela ne servira à rien

### La technique de Boolean Indexing, elle est très utile en Data Science et Machine Learning

```python
# Création d'un tableau rempli de nombre entiers aléatoires allant de 0 à 10 qui a pour dimension de (5, 5)
A = np.random.randint(0, 10, [5, 5])
A
```

```python
# Créer un masque rempli de True pour les éléments inférieurs à 5 et False pour ceux supérieurs à 5
A < 5
```

```python
# Remplir le tableau avec 11 où le masque est A < 5
A[A < 5] = 11
```

```python
A
```

```python
# On peut utiliser cela pour observer le comportement du tableau à un certain seuil et par la suite agir en conséquence
A[(A < 5) & (A > 2)] = 23
A
```

### Cette technique du Boolean Indexing est très utile pour le traitement d'image

```python
# 0 répresente un pixel noir et 255 un pixel blanc
image = np.random.randint(0, 255, [1024, 720])
image
```

```python
# Tous les pixels qui sont déjà supérieurs à 200 sont proches du blanc seront par défaut blanc, donc prende la valeur 255
image[image > 200] = 255
image
```

## Partie 2:
### Prendre une photo, la zoomer légèrement et appliquer un filtre dessus
## INDICES : 
### zoomer de 1/4 vers le milieu, utiliser les dimension de la photo avec le tuple shape, les enregistrer dans des variables et utiliser ces variables dans les techniques de slicing pour faire certaines choses comme zoomer de 1/4 vers le milieu de la photo. Ensuite augmenter la luminosité qui est déjà au dessus d'un certain seuil et aussi réduire la luminosité des pixels qui sont déjà très sombres et pour ça il faut utiliser le Boolean Indexing

```python
# Importation des modules nécessaires
from scipy import misc
import matplotlib.pyplot as plt

# Définition et visualisation de l'image
face = misc.face()
plt.imshow(face)
plt.show()

# face est un tableau numpy et on peut le vérifier
type(face)
```

```python
# Vérification des dimensions de face
face.shape
```

```python
# Nous sommes sur un tableau à 3D et donc il faut passer au tableau à 2D afin de bien travailler
face = misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray) # Si on enlève le dernier paramètre, l'image sera différente.
face.shape
```

```python
# On peut réduire de moitié le poids de cette image afin que l'espace qu'elle va occupée sur 
# Le disque dur ne soit pas aussi énorme
face = face[::2, ::2]
plt.imshow(face, cmap=plt.cm.gray) # Si on enlève le dernier paramètre, l'image sera différente.

# Affichage de la nouvelle taille de notre image
face.shape

# NB: Attention à ne pas trop réduire car on faisant ainsi votre image va perdre en qualité
```

```python
# Créer les variables contenant la hauteur la largeur
h = face.shape[0]
w = face.shape[1]

# Effectuer le zoom d'un quart sur toute l'image
# L'utilisation de h//4 ou w//4 nous permet d'avoir une division entière

zoom_image = face[h//4 : -h//4, w//4 : -w//4]
# Saturer les pixels plus proches du blanc avec une valeur de 255 en faisant le booleane indexing
zoom_image[zoom_image > 150] = 255
plt.imshow(zoom_image, cmap=plt.cm.gray)
plt.show()
```
