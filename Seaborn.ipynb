# Data Visualisation
##### Pour résoudre un problème en Data Science il faut comprendre les données, et pour ça il est mieux de les visualiser pour en tirer un max d'infos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data Set d'iris
iris = pd.read_csv("Iris.csv")
```

```python
iris.head()
```

```python
sns.pairplot(iris, hue="Species")
plt.show()
```

```python
# Dataset de Titanic
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.drop(["alone", "alive", "who", "adult_male", "embark_town", "class"], axis=1, inplace=True)
titanic.dropna(axis=0, inplace=True)
```

```python
titanic.head()
```

```python
# Visualiser pour voir à quoi ressemble ce data set
sns.pairplot(titanic)
plt.show()
```

##### Les box plot sont des graphiques qui nous permettent de voir la repartition des données avec des principales stats : médiane, quartile (minimun et maximum) sur différentes catégorie de variables 

### Catégorielle

```python
# C'est ce que l'on rencontre souvent lorsque l'on travaille sur les data set réels et il c'est alors 
# Qu'il faut utiliser l'affichage par type de données que l'on a en face de nous
sns.boxplot(x="pclass", y="age", data=titanic)
plt.show()
```

```python
# Afficher les classes des passagers en fonction de leurs l'âges
sns.catplot(x="survived", y="age", data=titanic, hue="sex")
plt.show()
```

### Distribution

```python
# La distribution de la variable "fare (tarif)"
sns.distplot(titanic["fare"])
plt.show()
```

```python
# L'âge en fonction des tarifs
sns.kdeplot(x="age", y="fare", data=titanic, hue="sex") # kde(Kernel Density Estimation)
plt.show()
```

```python
# L'âge en fonction des tarifs
sns.histplot(data=titanic, x="age", y="fare", hue="sex")
plt.show()
```

### Matrices

```python
# Matrice de corrélation
plt.figure(figsize=(10, 6))
sns.heatmap(titanic.corr())
plt.show()
```

## Seaborn vs Matplotlib
1. **Seaborn :**
    - Data (Dataset...)
    - Exploration Statistique
    - Vision Globale

2. **Matplotlib :**
    - Fonctions, Matrices, etc...
    - Mathématiques, Science, Ingénierie, etc...
    - Graphique Spécialisée

### Les fonctions les plus utiles avec Seaborn
- **Pairplot()**
- **Catplot()**
- **Boxplot()**
