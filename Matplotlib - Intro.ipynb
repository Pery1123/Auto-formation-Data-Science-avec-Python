```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
x = np.linspace(0, 2, 10)
y = x**2
print(x)
```

# MÃ©thode fonction

# 1) Pyplot

```python
# Il faut toujours avoir des tableaux de mÃªme dimensin de mÃªme taille sinon bonjour les dÃ©gats
plt.figure(figsize=(12, 8))
plt.plot(x, y, label="quadratique", lw=3, ls=":", c="g")
plt.plot(x, x**3, label="cubique", lw=3, ls="--", c="y")
plt.title("Figure_1")
plt.xlabel("Axe x")
plt.ylabel("Axe y")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(x, y, label="Ã©volution", lw=3, ls=":", c="g")
plt.scatter(np.sin(x), y, label="sinus de x", lw=3, ls="--", c="y")
plt.title("Figure_2")
plt.xlabel("Axe x")
plt.ylabel("Axe y")
plt.legend()
plt.show()
```

## SytÃ¨me de grille

```python
# DÃ©finition de la feuille blanche contenant toutes les informations des figures
plt.figure(figsize=(12, 8))

# Premier partie de la figure
plt.subplot(2, 2, 1)
plt.plot(x, y, label='quadratique', c='r')
plt.title("Figure_1")
plt.legend()

# DeuxiÃ¨me partie de la figure
plt.subplot(2, 2, 2)
plt.plot(x, np.sin(x), label="sinus ", c="y")
plt.plot(x, np.cos(x), label="cosinus", c='g')
plt.title("Figure_2")
plt.legend()


# TroisiÃ¨me partie de la figure
plt.subplot(2, 2, 3)
plt.plot(x, x**3, label="cubique")
plt.xlabel("L'axde des abscisses")
plt.ylabel("L'axe des ordonnÃ©es")
plt.title("Figure_3")
plt.legend()

# QuatriÃ¨me partie de la figure
plt.subplot(2, 2, 4)
plt.plot(x, np.sinh(x), label="sinus hyperbolique")
plt.plot(x, np.cosh(x), label="cosinus hyperbolique")
plt.title("FigureÃ§4")
plt.legend()

# Affichage de toutes les figures
plt.show()
```

# MÃ©thode OrientÃ©e Object

```python
# Avec cette mÃ©thode on peut avoir des figures qui partagent le mÃªme axe
# fig est un objet, ax est un tableau numpy contenant des objets
fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(x, y, label="consinus")
#plt.legend()
ax[1].plot(x, np.tan(x), label="tangente")
plt.legend()
plt.show()
```

## Exercice
#### Vous avez un dictionnaire qui comprend ***N datasets***. CrÃ©ez une fonction qui va tracer ***N graphiques*** sur une seule et mÃªme figure

```python
dataset = {f"experience{i}": np.random.randn(100, 3) for i in range(6)}
#dataset
```

```python
# CrÃ©ation de la fonction graphique
def graphique(data):
    
    # Prendre la taille du dataset
    n = len(data)
    # CrÃ©ation de la figure
    plt.figure(figsize=(12, 30))
    
    for k, i in zip(data.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(data[k])
        plt.title(k)
    plt.show()
```

```python
graphique(dataset)
```

```python
# Charger les donnÃ©es de la plante iris dans un tableau
iris = load_iris()

# Enregistrer les donnÃ©es et les cibles(ici les classes) dans les tableaux x et y
x = iris.data
y = iris.target
names = list(iris.target_names)

print(f"x contient {x.shape[0]} exemples et {x.shape[1]} variables")
print(f"il y a {np.unique(y).size} classes")
```

## Graphique de Classification avec plt.scatter()

```python
# Afficher toutes les lignes des deux 1eres variables en fonction de la couleur de chaque classe

# alpha(pour controler la transparence des points)
# s(contrÃ´le la taille des poids)
plt.figure(figsize=(10, 6))
plt.scatter(x[:, 0], x[:, 1], c=y, alpha=1, s=x[:, 2]*100)
plt.xlabel("Longeur de sÃ©pal")
plt.ylabel("Largeur de sÃ©pal")
plt.legend()
plt.show()
```

```python
# Visualisation des fonctions
f = lambda x, y: np.sin(x) + np.cos(x+y)*np.cos(x)


X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
# CrÃ©ation de la gride de travail
X, Y = np.meshgrid(X, Y)

# Injecter X et Y dans la fonction crÃ©Ã©e
Z = f(X, Y)

# VÃ©rification de la taille de Z
Z.shape
```

```python
# Visualisation
# Feuille de rÃ©presentation
plt.figure(figsize=(12, 8))
# CrÃ©ation de l'object ax en projection 3D
ax = plt.axes(projection="3d")
ax.plot_surface(X, Y, Z, cmap="plasma")
plt.show()
```
